# Phase 6 — Experiments and Visualizations

Consolidates the seven definitive experiments. Each cell shows the figure produced by the corresponding script in `scripts/`. To regenerate every figure run:

```bash
python scripts/run_all_experiments.py
```

````python
from pathlib import Path
from IPython.display import Image, Markdown, display

FIG = Path('../figures')

EXPERIMENTS = [
    ('A', 'Transition convergence', 'transition_convergence_heatmap.png',
     'Validates the convergence theorem $P^n \\to \\mathbf{1}\\pi$ from Phase 2.'),
    ('B', 'Eigenvalue spectrum', 'eigenvalue_spectrum.png',
     'Spectrum of $P$ in the unit disk; spectral gap $\\gamma = 1 - |\\lambda_2|$ controls mixing.'),
    ('C', 'Absorption times', 'absorption_times.png',
     'Expected months to exit per starting level under the absorbing chain (Phase 3).'),
    ('D₁', 'Birth-death stationary (analytical)', 'birth_death_stationary.png',
     'Detailed-balance $\\pi_n$ overlaid with Poisson($\\rho$) — Phase 4 §3.'),
    ('D₂', 'Birth-death stationary (simulation)', 'birth_death_steady_simulation.png',
     'Empirical histogram of simulated team sizes after burn-in vs Poisson($\\rho$).'),
    ('E', 'Headcount scenarios', None,
     'Five scenarios: steady growth, hiring freeze, layoff recovery, composition shift, seasonal hiring.'),
    ('F', 'Sensitivity tornado', 'sensitivity_tornado.png',
     'Rate-by-rate impact of ±50% perturbations on Senior share and steady-state mean team size.'),
    ('G', 'Trajectory animation', 'trajectory_animation.gif',
     '100 simulated trajectories revealed month-by-month with the analytical mean and 5–95% band.'),
]

for exp_id, name, fig_name, description in EXPERIMENTS:
    display(Markdown(f'## Experiment {exp_id} — {name}'))
    display(Markdown(description))
    if fig_name is None:
        for s in ['scenario_s1_steady_growth.png', 'scenario_s2_hiring_freeze.png',
                  'scenario_s3_layoff_recovery.png', 'scenario_s4_composition_shift.png',
                  'scenario_s5_seasonal_hiring.png']:
            display(Image(filename=str(FIG / s)))
    else:
        display(Image(filename=str(FIG / fig_name)))
````
